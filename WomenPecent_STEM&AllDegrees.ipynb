# Gender Gap in College Major

The goal of this analysis is to provide answer for the following questions:
+ Do percent of women varies within different majors
+ Do percent of women gap changed over time

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

women_degree = pd.read_csv ('percent-bachelors-degrees-women-usa.csv')
women_degree.head(1)
```

Let use classfiy majors into STEM, Liberal Art, and Other catagories

## Comparing across all major degrees

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics',
             'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 
                 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 
              'Agriculture','Business', 'Architecture']
```

```python
fig = plt.figure(figsize = (16, 20))

# Generate plot for  STEM degrees in the first column
for sp in range (0, 18, 3):
    act_index = int(sp/3)
    ax = fig.add_subplot(6, 3, sp+1)
    ax.plot(women_degree['Year'],women_degree[stem_cats[act_index]], c = cb_dark_blue, label = 'Women', linewidth = 3)
    ax.plot(women_degree['Year'],(100-women_degree[stem_cats[act_index]]), c = cb_orange, label = 'Men', linewidth = 3)
    ax.set_title(stem_cats[act_index])
    for key, spine in ax.spines.items(): # remove spines from upper, bottom, left and right
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c = (171/255, 171/255, 171/255), alpha = 0.3)# add horizontal line for better visualization
    ax.tick_params (bottom = 'off', top = 'off', right = 'off', left = 'off', labelbottom = 'off')
    if sp == 0: # add text anotation for the first and last plot in the first column
        ax.text(2003, 87, 'Women')
        ax.text(2002, 10, 'Men')
    elif sp == 15:
        ax.text(2005, 5, 'Women')
        ax.text(2001, 95, 'Men')
        ax.tick_params(labelbottom = 'on')
        
# Generate plot for  Liberal Arts in the second column
for sp in range (1, 16, 3):
    act_index = int((sp-1)/3)
    ax = fig.add_subplot(6, 3, sp+1)
    ax.plot(women_degree['Year'],women_degree[lib_arts_cats[act_index]], c = cb_dark_blue, label = 'Women', linewidth = 3)
    ax.plot(women_degree['Year'],(100-women_degree[lib_arts_cats[act_index]]), c = cb_orange, label = 'Men', linewidth = 3)
    ax.set_title(lib_arts_cats[act_index])
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.axhline(50, c = (171/255, 171/255, 171/255), alpha = 0.3)
    ax.set_yticks([0,100])# reduce clutter from y axis
    ax.set_ylim(0,100)
    ax.tick_params (bottom = 'off', top = 'off', right = 'off', left = 'off', labelbottom = 'off')
    if sp == 0:
        ax.text(2003, 87, 'Women')
        ax.text(2002, 10, 'Men')
    elif act_index == 4:
        ax.tick_params(labelbottom='on')
 
# Generate plot for  Liberal Arts in the second column
for sp in range (2, 20, 3):
    act_index = int((sp-2)/3)
    ax = fig.add_subplot(6, 3, sp+1)
    ax.plot(women_degree['Year'],women_degree[other_cats[act_index]], c = cb_dark_blue, label = 'Women', linewidth = 3)
    ax.plot(women_degree['Year'],(100-women_degree[other_cats[act_index]]), c = cb_orange, label = 'Men', linewidth = 3)
    ax.set_title(other_cats[act_index])
    ax.set_yticks([0,100])
    ax.axhline(50, c =(171/255, 171/255, 171/255), alpha = 0.3)
    for key, spine in ax.spines.items(): 
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.tick_params (bottom = 'off', top = 'off', right = 'off', left = 'off', labelbottom = 'off')
    
    if sp == 2:
        ax.text(2003, 92, 'Women')
        ax.text(2002, 5, 'Men')
        
    elif act_index == 5:
        ax.text(2005, 30, 'Women')
        ax.text(2001, 65, 'Men')
        ax.tick_params(labelbottom = 'on')
plt.savefig("gender_degrees.PDF")
plt.show
```
